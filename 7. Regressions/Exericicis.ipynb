Inicialment hem importat la llibreria pandas per llegir el dataset i hem utilitzat les funcions head(), info(), describe() i display() per tenir una visió general sobre la informació del dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
dataset = pd.read_csv('./test-data.csv')

#display(dataset.head())
#print(dataset.info())
#display(dataset.describe())
#display(dataset.isnull().sum())
# Comentar
dataset.drop(columns=['New_Price'], inplace=True)
# Comentar
dataset['Power'] = dataset['Power'].replace('null bhp', np.nan)
dataset['Power'] = dataset['Power'].str.replace(' bhp', '', regex=False)
dataset['Power'] = dataset['Power'].astype(float)

dataset['Power'].fillna(dataset['Power'].mode()[0], inplace=True)

dataset['Engine'] = dataset['Engine'].str.replace(' CC', '').astype(float)
dataset['Mileage'] = dataset['Mileage'].str.replace(' kmpl', '').str.replace(' km/kg', '').astype(float)
#
dataset['Name'] = dataset['Name'].str.split().str[0]
#
dataset = pd.get_dummies(dataset, columns=['Fuel_Type', 'Transmission', 'Owner_Type', 'Name', 'Location'])
bool_cols = dataset.select_dtypes(include='bool').columns
dataset[bool_cols] = dataset[bool_cols].astype(int)
display(dataset.head())
print(dataset.info())
display(dataset.describe())

numerical_features = dataset.select_dtypes(include=['float64', 'int64'])
correlation = numerical_features.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm')
plt.show()
correlation_price = correlation['Price'].sort_values(ascending=False)
print(correlation_price)

```
